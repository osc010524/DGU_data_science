# 보스턴 집값 예측 속성을 8개로 줄이기

```python
import tensorflow as tf
import pandas as pd
```

```python
file_path = "./data/boston.csv"
bost = pd.read_csv(file_path)
bost.columns
```

```python
indepan = bost[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis']]
depen = bost[['medv']]
print(indepan.shape, depen.shape)
```

```python
# 모델 구조 생성
X = tf.keras.layers.Input(shape=[8])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```python
# 모델 학습
model.fit(indepan, depen, epochs=1000)
```

```python
# 모델 이용
model.predict(indepan[0:5])
```
