```python
import pandas as pd
```

# 파일 수집

```python
file_path = 'data/boston.csv'
```

```python
bost = pd.read_csv(file_path)
print(bost.columns)
```

```python
bost.head()
```

```python
bost.tail()
```

```python
# 독립 = independent, 종속 = dependent
# 속성은 대소문자 구분하지 않음
indepen = bost[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'b', 'lstat']]
depen = bost[['medv']]
print(indepen.shape, depen.shape)
```

```python
# 모델의 구조 새성
import tensorflow as tf
x = tf.keras.layers.Input(shape=[13])
y = tf.keras.layers.Dense(1)(x)
model = tf.keras.models.Model(x, y)
model.compile(loss='mse')
```

```python
# 모델 학습
model.fit(indepen, depen, epochs=1000)
```

```python
# 모델 이용
model.predict(indepen[0:5])
```

```python
# 실제 csv 값의 종속값
depen[0:5]
```

```python
# 모델 수식 확인
model.get_weights()
```
